*  # PROBLEMATICA
    La movilidad social es un fenomeno de índole económica y social ya que se refiere a la posibilidad de un individuo (o sociedad) de algún estrato socieconomico cambiar a otro. Y es de suma importancia, ya que nos da un punto de referncia para saber que tanto una sociedad se vuelve más equtativa y que tantas oportunidades suelen tener para llegar a una mejor calidad de vida.
    Existen dos tipos, la **movilidad social absoluta (intrageneracional)** y la **movilidad social relativa (intergeneraicional)**. La primera se encarga de medir los cambios de un individuo a lo largo de su ciclo de vida. La segunda a la que nos aremos enfasis en este trabajo, la movilidad Intergeneracional mide los cambios entre generaciones de una misma familia.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
variables_esru=pd.read_csv('/content/drive/Othercomputers/Mi portátil/Escritorio/DIPLOMADO UNAM CIENCIA DE DATOS/MOVILIDAD SOCIAL/PROYECTO FINAL/variables_esru.csv')
```

```python
columnas_analizar=['folio','Estado','p05','p06','p133','region','cdmx','educ',
       'educ_padre', 'educ_madre', 'educ_padre_2','i_or',
       'iriq_or', 'quintil_or', 'i_des', 'iriq_des', 'quintil_des',
       'clases_padre', 'clases_madre', 'clases', 'niveducp', 'niveducm',
       'niveduc', 'anesc_pad', 'anesc_mad', 'anesc_padres', 'anesc_ent',
       'padres_hac', 'hijo_hac','riq_padres',
       'q_riq_padres', 'riq_hijo', 'q_riq_hijo', 'hogar_actual',
       'hogar_nin', 'pobre', 'region_14', 'cdmx_14']
movilidad_s=pd.DataFrame(variables_esru[columnas_analizar])
print(movilidad_s)
```

```python

region_leyenda=['Centro-Norte','Norte','Norte-Occidente','Sur','Centro']
etiquetas=['0','Nula (E)','Primaria Incompleta (D-)','Primaria (D)','Secundaria (C-/C)','Preparatoria (C+)','Profesional (A/B)']


colores = [
    'rgba(255, 255, 0, 1)',
    'rgba(255, 0, 0, 1)',
    'rgba(0, 0, 255, 1)',
    'rgba(0, 255, 0, 1)',
    'rgba(255, 165, 0, 1)'
]

fig1 = px.histogram(movilidad_s, x='educ', color='region', title='Educación del Entrevistado',
                    color_discrete_sequence=colores)
for i, region_name in enumerate(region_leyenda):
    fig1.data[i].name = region_name
fig1.update_layout(barmode='group', xaxis_title='Educación', yaxis_title='Conteo',
                  xaxis=dict(ticktext=etiquetas, tickvals=list(range(7))))


fig2 = px.histogram(movilidad_s, x='educ_padre', color='region', title='Educación del Padre',
                    color_discrete_sequence=colores)
for i, region_name in enumerate(region_leyenda):
    fig2.data[i].name = region_name
fig2.update_layout(barmode='group', xaxis_title='Educación del Padre', yaxis_title='Conteo',
                   xaxis=dict(ticktext=etiquetas, tickvals=list(range(7))))


fig3 = px.histogram(movilidad_s, x='educ_madre', color='region', title='Educación de la Madre',
                    color_discrete_sequence=colores)
for i, region_name in enumerate(region_leyenda):
    fig3.data[i].name = region_name
fig3.update_layout(barmode='group', xaxis_title='Educación de la Madre', yaxis_title='Conteo',
                   xaxis=dict(ticktext=etiquetas, tickvals=list(range(7))))


fig1.show()
fig2.show()
fig3.show()
```

```python
etiquetas_riq=['Quintil 1','Quintil 2','Quintil 3','Quintil 4','Quintil 5']
fig13 = px.line(movilidad_s, x='quintil_or', y='quintil_des', title='Riqueza por Quintiles')
fig13.update_layout(xaxis_title='Quintil de Riqueza de Origen', yaxis_title='Quintil de Riqueza Destino',
                   xaxis=dict(ticktext=etiquetas_riq, tickvals=list(range(5))),
                   yaxis=dict(ticktext=etiquetas_riq, tickvals=list(range(5))))

fig13.show()
```

```python
etiquetas_riq=['Quintil 1','Quintil 2','Quintil 3','Quintil 4','Quintil 5']

conteo = movilidad_s.groupby(['quintil_or', 'quintil_des']).size().reset_index(name='conteo')
fig14 = px.line(conteo, x='quintil_or', y='quintil_des', color='conteo', markers=True, title='Conteo de Movilidad en Riqueza')
fig14.update_layout(xaxis_title='Quintil de Riqueza de Origen', yaxis_title='Quintil de Riqueza Destino',
                   xaxis=dict(ticktext=etiquetas_riq, tickvals=list(range(5))),
                   yaxis=dict(ticktext=etiquetas_riq, tickvals=list(range(5))))
fig14.show()
```

```python
correlation_matrix = movilidad_s[['riq_padres', 'riq_hijo', 'educ', 'educ_padre_2']].corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlaciones Riqueza y Educación')
plt.show()
```
