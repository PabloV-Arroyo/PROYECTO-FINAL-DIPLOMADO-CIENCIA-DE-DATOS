```python
#INDICE DE RIQUEZA DE HOGAR DE ORIGEN
!pip install prince
import pandas as pd
from prince import MCA

esru['piso'] = np.nan
esru.loc[esru['p29'].isin([2, 3]), 'piso'] = 1
esru.loc[esru['p29'] == 1, 'piso'] = 0

esru['piso'] = esru['piso'].astype('category')
unique_categories = {0: 'Piso de cemento', 1: 'Piso de madera', 2: 'Otro tipo de piso'}
esru['piso'].cat.rename_categories(unique_categories)

for col in esru[['p30_a', 'p30_b', 'p30_c', 'p30_d', 'p30_e', 'p33_a', 'p33_b', 'p33_c', 'p33_d', 'p33_e', 'p33_f', 'p33_g', 'p33_h', 'p33_i', 'p33_j', 'p33_k', 'p33_l', 'p33_m', 'p33_n']].columns:
    esru[col] = esru[col].replace({2: 0, 8: 0})
```

```python
mca = MCA(n_components=2)
mca = mca.fit(esru[['p30_a', 'p30_b', 'p30_c', 'p30_d', 'p30_e', 'p33_a', 'p33_b', 'p33_c', 'p33_d', 'p33_e', 'p33_f', 'p33_g', 'p33_h', 'p33_i', 'p33_j', 'p33_k', 'p33_l', 'p33_m', 'p33_n']])


transformed_data = mca.transform(esru[['p30_a', 'p30_b', 'p30_c', 'p30_d', 'p30_e', 'p33_a', 'p33_b', 'p33_c', 'p33_d', 'p33_e', 'p33_f', 'p33_g', 'p33_h', 'p33_i', 'p33_j', 'p33_k', 'p33_l', 'p33_m', 'p33_n']])

esru['i_or'] = transformed_data.iloc[:, 0]
esru['i_or'] = esru['i_or'] * -1


mean_i_or = esru['i_or'].mean()
std_i_or = esru['i_or'].std()


esru['iriq_or'] = (esru['i_or'] - mean_i_or) / std_i_or


esru['iriq_or'] = esru['iriq_or'].astype('category')
esru['iriq_or'].cat.rename_categories({0: 'Índice de riqueza del hogar de origen'})

esru['iriq_or'] = esru['iriq_or'].astype(float)

esru['quintil_or'] = pd.qcut(esru['iriq_or'], 5, labels=False)
```
